```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('basics').getOrCreate()
```

```python
df = spark.read.csv('Datasets/Accidents_2015.csv',header=True,inferSchema=True)
```

```python

date_time = df.select('Day_of_Week', df['Time'].substr(0, 2).alias("Time"))
```

```python
date_time.show()
```

```python
test = date_time.toPandas()  
```

```python
test
```

```python
test["Day_of_Week"] = test["Day_of_Week"].astype(str)
```

```python
test['Day'] = test["Day_of_Week"]
test.head()
```

```python
test["Day_of_Week"] = test["Day_of_Week"].replace("1", "Sunday")
test["Day_of_Week"] = test["Day_of_Week"].replace("2", "Monday")
test["Day_of_Week"] = test["Day_of_Week"].replace("3", "Tuesday")
test["Day_of_Week"] = test["Day_of_Week"].replace("4", "Wednesday")
test["Day_of_Week"] = test["Day_of_Week"].replace("5", "Thursday")
test["Day_of_Week"] = test["Day_of_Week"].replace("6", "Friday")
test["Day_of_Week"] = test["Day_of_Week"].replace("7", "Saturday")

test.head()
```
